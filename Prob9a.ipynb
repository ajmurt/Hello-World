```python
from collections import Counter

def readlines_txt(fname):
    with open(fname) as f:
        lines = f.read().splitlines()  
#    for line in lines:
#        print(line)
    return lines

filename = 'prob9a_data.txt'
#filename = 'Prob9a_testdata.txt'
stuff = readlines_txt(filename)
print(len(stuff), 'lines in data file', '\n')


for s in stuff[:20]:
    print(s)

lines = []
for s in stuff:    
    lines.append( [int(i) for i in s.split(' ')] )

print('\n')
for l in lines[:20]:
    print(l)
```

```python
def reduce_lst(lst):
    red_lst = [ lst[j]-lst[j-1] for j in range(1, len(lst)) ]
    return red_lst

def all_zero(mylist):
    flag = len(set(mylist)) == 1 and 0 in set(mylist)
    return flag

```

```python
def get_to_zero(l):

    i = 0
    imax = 100
    next_in_seq = l[-1]
    done = False
    while not done:
        i +=  1
        l = reduce_lst(l)
        next_in_seq += l[-1]
        if all_zero(l):
            done = True
        if i > 30:
            done = True
            print("reached imax")
    return next_in_seq
```

```python
answer = 0
for l in lines:
    l.reverse()
    next = get_to_zero(l)
    answer += next
#    print(next, answer)
    
print('\n', "answer is", answer)
```
